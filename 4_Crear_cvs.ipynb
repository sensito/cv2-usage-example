```python
import glob

import json

import numpy as np

import pandas as pd
from pandas.io.json import json_normalize  
```

```python
filenames = glob.glob("./JSON_Caras_Partes/**")
```

```python
filenames
```

```python
clientes = pd.DataFrame(dtype=np.int8)
```

```python
    DataSet =  json_normalize(data["Person"][1]["DataSet"]) 
```

```python
for key in range(0,len(filenames)):
    with open('./JSON_Caras_Partes/data_' + str(key) + '.json', 'r') as file:
        data = json.load(file)
    Persona = json_normalize(data["Person"][0]) 
    DataSet =  json_normalize(data["Person"][1]["DataSet"]) 
    clients = pd.concat([Persona, DataSet], axis=1)
    clientes = clientes.append(clients)
```

```python
clientes.reset_index(drop=True, inplace=True)
```

```python
clientes
```

```python
clientes.to_csv('cara.csv')
```
